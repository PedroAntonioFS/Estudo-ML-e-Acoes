```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('transformation_result_1_year_SD.csv')
```

```python
df.head()
```

```python
df['Open_dif'] = df['Open_1'] - df['Open_0']
df.drop(['Open_1', 'Open_0'], axis=1, inplace=True)
```

```python
for col in df.drop('Open_dif', axis=1).columns:
    x = df[[col, 'Open_dif']]
    km = KMeans(n_clusters=2)
    
    y = km.fit_predict(x)
    
    plt.scatter(
        x['Open_dif'][y == 0], x[col][y == 0],
        s=50, c='lightgreen',
        marker='s', edgecolor='black',
        label='cluster 1'
    )

    plt.scatter(
        x['Open_dif'][y == 1], x[col][y == 1],
        s=50, c='orange',
        marker='o', edgecolor='black',
        label='cluster 2'
    )
    
    plt.title("{} x Price in a Year".format(col))
    plt.xlabel('Price')
    plt.ylabel(col)
    plt.legend(scatterpoints=1)
    plt.grid()
    plt.show()
    print("cluster 1:{} / Cluster 2:{}\n".format(len(x[y == 0]), len(x[y == 1])))
```

```python
for col in df.drop('Open_dif', axis=1).columns:
    x = df[[col, 'Open_dif']]
    km = KMeans(n_clusters=3)
    
    y = km.fit_predict(x)
    
    plt.scatter(
        x['Open_dif'][y == 0], x[col][y == 0],
        s=50, c='lightgreen',
        marker='s', edgecolor='black',
        label='cluster 1'
    )

    plt.scatter(
        x['Open_dif'][y == 1], x[col][y == 1],
        s=50, c='orange',
        marker='o', edgecolor='black',
        label='cluster 2'
    )

    plt.scatter(
        x['Open_dif'][y == 2], x[col][y == 2],
        s=50, c='blue',
        marker='o', edgecolor='black',
        label='cluster 3'
    )
    
    plt.title("{} x Price in a Year".format(col))
    plt.xlabel('Price')
    plt.ylabel(col)
    plt.legend(scatterpoints=1)
    plt.grid()
    plt.show()
    print("cluster 1:{} / Cluster 2:{}\n / Cluster 3:{}\n".format(len(x[y == 0]), len(x[y == 1]), len(x[y == 2])))
```
