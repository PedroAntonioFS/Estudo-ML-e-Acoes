```python
import numpy as np
import pandas as pd
```

```python
fundamental_data = pd.read_csv('preprocessing_result_DF.csv')
prices_data = pd.read_csv('preprocessing_result_CH.csv')
prices_data['Date'] = pd.to_datetime(prices_data['Date'])
```

```python
fundamental_data.head()
```

```python
prices_data.head()
```

## 1. Remover registros de dados fundamentalistas de 2022
Será utilizado a cotação de ,no mínimo, um ano depois de cada registro de dados fundamentalista. Será removido os registros de 2022 porque não cotação de 2023 ou mais.

```python
fundamental_data.where(fundamental_data['Ano'] != 2022, inplace=True)
fundamental_data.dropna(inplace=True)
fundamental_data.reset_index(drop=True, inplace=True)
```

```python
fundamental_data.head()
```

## 2. Mesclar a duas bases de dados

```python
fundamental_data.shape
```

```python
prices_data['Year'] = prices_data['Date'].dt.year
prices_data['Month'] = prices_data['Date'].dt.month
```

```python
fundamental_data.rename(columns = {'Ano': 'Year'}, inplace=True)
```

```python
stocks_data = fundamental_data.merge(prices_data, on=['Ticker', 'Year', 'Tipo']).drop(['Date', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1)
```

```python
stocks_data.head()
```

## 3. Calcular atributos alvo da aplicação das técnicas de apredizado de máquina
Foi calculado o preço em 1,2,3,4,5 e 10 anos.

```python
prices_after_1_year = prices_data.copy()
prices_after_2_years = prices_data.copy()
prices_after_3_years = prices_data.copy()
prices_after_4_years = prices_data.copy()
prices_after_5_years = prices_data.copy()
prices_after_10_years = prices_data.copy()
```

```python
prices_after_1_year['Year'] = prices_after_1_year['Year'] - 1
prices_after_2_years['Year'] = prices_after_2_years['Year'] - 2
prices_after_3_years['Year'] = prices_after_3_years['Year'] - 3
prices_after_4_years['Year'] = prices_after_4_years['Year'] - 4
prices_after_5_years['Year'] = prices_after_5_years['Year'] - 5
prices_after_10_years['Year'] = prices_after_10_years['Year'] - 10
```

```python
stock_data_1_year = stocks_data.merge(prices_after_1_year, on=['Ticker', 'Year', 'Month', 'Tipo'], suffixes=('_0', '_1')).drop(['Date', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1)
stock_data_2_years = stocks_data.merge(prices_after_2_years, on=['Ticker', 'Year', 'Month', 'Tipo'], suffixes=('_0', '_1')).drop(['Date', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1)
stock_data_3_years = stocks_data.merge(prices_after_3_years, on=['Ticker', 'Year', 'Month', 'Tipo'], suffixes=('_0', '_1')).drop(['Date', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1)
stock_data_4_years = stocks_data.merge(prices_after_4_years, on=['Ticker', 'Year', 'Month', 'Tipo'], suffixes=('_0', '_1')).drop(['Date', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1)
stock_data_5_years = stocks_data.merge(prices_after_5_years, on=['Ticker', 'Year', 'Month', 'Tipo'], suffixes=('_0', '_1')).drop(['Date', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1)
stock_data_10_years = stocks_data.merge(prices_after_10_years, on=['Ticker', 'Year', 'Month', 'Tipo'], suffixes=('_0', '_1')).drop(['Date', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], axis=1)
```

```python
stock_data_1_year.head()
```

## 4. Manter Apenas 1 registro por ano e ticker, selecionando o primeiro valor do ano

```python
first_month_1_year = stock_data_1_year[['Year', 'Ticker', 'Tipo', 'Month']].groupby(['Year', 'Ticker', 'Tipo']).min()
first_month_2_years = stock_data_2_years[['Year', 'Ticker', 'Tipo', 'Month']].groupby(['Year', 'Ticker', 'Tipo']).min()
first_month_3_years = stock_data_3_years[['Year', 'Ticker', 'Tipo', 'Month']].groupby(['Year', 'Ticker', 'Tipo']).min()
first_month_4_years = stock_data_4_years[['Year', 'Ticker', 'Tipo', 'Month']].groupby(['Year', 'Ticker', 'Tipo']).min()
first_month_5_years = stock_data_5_years[['Year', 'Ticker', 'Tipo', 'Month']].groupby(['Year', 'Ticker', 'Tipo']).min()
first_month_10_years = stock_data_10_years[['Year', 'Ticker', 'Tipo', 'Month']].groupby(['Year', 'Ticker', 'Tipo']).min()
```

```python
first_month_1_year.head()
```

```python
def is_min(row, group_by):
    return row['Month'] == group_by.loc[row['Year'], row['Ticker'], row['Tipo']]['Month']
```

```python
stock_data_1_year = stock_data_1_year[stock_data_1_year.apply(lambda x: is_min(x, first_month_1_year), axis=1)].reset_index(drop=True)
stock_data_2_years = stock_data_2_years[stock_data_2_years.apply(lambda x: is_min(x, first_month_2_years), axis=1)].reset_index(drop=True)
stock_data_3_years = stock_data_3_years[stock_data_3_years.apply(lambda x: is_min(x, first_month_3_years), axis=1)].reset_index(drop=True)
stock_data_4_years = stock_data_4_years[stock_data_4_years.apply(lambda x: is_min(x, first_month_4_years), axis=1)].reset_index(drop=True)
stock_data_5_years = stock_data_5_years[stock_data_5_years.apply(lambda x: is_min(x, first_month_5_years), axis=1)].reset_index(drop=True)
stock_data_10_years = stock_data_10_years[stock_data_10_years.apply(lambda x: is_min(x, first_month_10_years), axis=1)].reset_index(drop=True)
```

```python
stock_data_1_year.head()
```

```python
# stock_data_1_year['Open_dif'] = stock_data_1_year['Open_1'] - stock_data_1_year['Open_0']
# stock_data_2_years['Open_dif'] = stock_data_2_years['Open_1'] - stock_data_2_years['Open_0']
# stock_data_3_years['Open_dif'] = stock_data_3_years['Open_1'] - stock_data_3_years['Open_0']
# stock_data_4_years['Open_dif'] = stock_data_4_years['Open_1'] - stock_data_4_years['Open_0']
# stock_data_5_years['Open_dif'] = stock_data_5_years['Open_1'] - stock_data_5_years['Open_0']
```

```python
stock_data_1_year['years_dif'] = 1
stock_data_2_years['years_dif'] = 2
stock_data_3_years['years_dif'] = 3
stock_data_4_years['years_dif'] = 4
stock_data_5_years['years_dif'] = 5
```

## 5. Mesclar todas as bases de dados em uma única

```python
stock_data = pd.concat([stock_data_1_year, stock_data_2_years, stock_data_3_years, stock_data_4_years, stock_data_5_years])
```

## 6. Remover "Year", "Month" e "Ticker"

```python
stock_data_1_year.drop(['Year', 'Month', 'Ticker', 'years_dif'], axis=1, inplace=True)
stock_data_2_years.drop(['Year', 'Month', 'Ticker', 'years_dif'], axis=1, inplace=True)
stock_data_3_years.drop(['Year', 'Month', 'Ticker', 'years_dif'], axis=1, inplace=True)
stock_data_4_years.drop(['Year', 'Month', 'Ticker', 'years_dif'], axis=1, inplace=True)
stock_data_5_years.drop(['Year', 'Month', 'Ticker', 'years_dif'], axis=1, inplace=True)
stock_data.drop(['Year', 'Month', 'Ticker'], axis=1, inplace=True)
```

## 7. Mapear coluna "Tipo" para inteiro

```python
stock_data_1_year = stock_data_1_year.join(pd.get_dummies(stock_data_1_year['Tipo'])).drop('Tipo', axis=1)
stock_data_2_years = stock_data_2_years.join(pd.get_dummies(stock_data_2_years['Tipo'])).drop('Tipo', axis=1)
stock_data_3_years = stock_data_3_years.join(pd.get_dummies(stock_data_3_years['Tipo'])).drop('Tipo', axis=1)
stock_data_4_years = stock_data_4_years.join(pd.get_dummies(stock_data_4_years['Tipo'])).drop('Tipo', axis=1)
stock_data_5_years = stock_data_5_years.join(pd.get_dummies(stock_data_5_years['Tipo'])).drop('Tipo', axis=1)
stock_data = stock_data.join(pd.get_dummies(stock_data['Tipo'])).drop('Tipo', axis=1)
```
