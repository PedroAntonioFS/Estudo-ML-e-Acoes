```python
import pandas as pd
import numpy as np
import re
```

# 1. Primeira base de dados
## 1.1. Identificando problemas na base de dados
Os problemas identificados foram:
1. Existem registros com valores vazios;
2. Números não estão com tipo correto (float ou int).

```python
fundamental_data = pd.read_csv('dados_fudamentalistas.csv')
```

```python
fundamental_data.head()
```

```python
fundamental_data.dtypes
```

```python
fundamental_data['Ano'].unique()
```

```python
unique_vals = set()
for col in fundamental_data.columns:
    if col in ['Ano', 'Ticker', 'Tipo']:
        continue
    
    for val in fundamental_data[col]:
        if not re.search("^-?[0-9]*\,[0-9]*\%?$", val):
            unique_vals.add(val)
```

```python
unique_vals
```

## 1.2. Remover caracteres '%' e '.'

```python
fundamental_data = fundamental_data.applymap(lambda x: x.replace('%',''))
```

```python
fundamental_data = fundamental_data.applymap(lambda x: x.replace('.',''))
```

```python
fundamental_data.head()
```

## 1.3. Substituir caracteres ',' por '.'

```python
fundamental_data = fundamental_data.applymap(lambda x: x.replace(',','.'))
```

```python
fundamental_data.head()
```

## 1.4. Mapear valores ausente '-' para np.nan

```python
fundamental_data.replace("-", np.NAN, inplace=True)
```

```python
fundamental_data.head()
```

## 1.5. Mapear valores "ATUAL" para "2022"

```python
fundamental_data.replace("ATUAL", "2022", inplace=True)
```

```python
fundamental_data.head()
```

## 1.6. Converter números do tipo string para float ou int

```python
for col in fundamental_data.columns:
    if col in ['Ticker', 'Tipo']:
        continue
        
    fundamental_data[col] = pd.to_numeric(fundamental_data[col])
```

```python
fundamental_data.head()
```

```python
fundamental_data.dtypes
```

## 1.7. Lidar com valores ausentes

```python
fundamental_data.shape
```

```python
fundamental_data.isnull().sum()/fundamental_data.shape[0]
```

### 1.7.1. Foi aplicado:
1. Remoção dos registros com mais de 50% de valores nulos;
2. Remoção das colunas com mais de 50% dae valores nulos;
3. Preenchimento dos demais valores nulos com a média de cada coluna.

```python
fundamental_data.drop((fundamental_data.isnull().sum(axis=1)/fundamental_data.shape[1])[lambda x: x > 0.5].index, inplace=True)
```

```python
fundamental_data.drop((fundamental_data.isnull().sum()/fundamental_data.shape[0])[lambda x: x > 0.5].index, axis=1, inplace=True)
```

```python
fundamental_data.shape
```

```python
for col in fundamental_data.columns:
    if col in ['Ano', 'Ticker', 'Tipo']:
        continue
    
    col_mean = fundamental_data[col].mean()
    fundamental_data[col].fillna(value=col_mean, inplace=True)
```

```python
fundamental_data.isnull().sum()/fundamental_data.shape[0]
```

```python
fundamental_data.reset_index(drop=True, inplace=True)
```

```python
fundamental_data.to_csv('preprocessing_result_DF.csv', index=False)
```
