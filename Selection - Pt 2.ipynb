### 2. Primeira Iteração de coleta de dados do website Yahoo Finance

Coleta as cotações históricas da lista de ativos de df e armazena em prices_data. Os casos com erros (stock_errors) são mapeados em df, na coluna 'error'.

```python
scraper = YahooFinanceScraper()
```

```python
from_date = datetime(2011, 1, 1)
to_date = datetime(2022, 6, 1)
interval = '1mo'
```

```python
stock_errors = []
prices_data = None

for idx, data in df.iterrows():
    ticker = data['Ativos'].strip()
    try:
        stock_data = scraper.get_data(data['Tipo'], ticker, from_date, to_date, interval)
        stock_data['Ticker'] = data['Ativos']
        stock_data['Tipo'] = data['Tipo']
        if prices_data is None:
            prices_data = stock_data
        else:
            prices_data = pd.concat([prices_data, stock_data])
        print("log {}/{}: Success".format(idx,df.shape[0]))
    except (IOError, ExceptionsFacade.get_NoSuchElementException(), ExceptionsFacade.get_TimeoutException()) as er:
        stock_errors.append(idx)
        print("log {}/{}: Error -> {}".format(idx,df.shape[0], scraper.url))
df['error'] = 0
df['error'].iloc[stock_errors] = 1
```

### 3. Segunda Iteração de coleta de dados do website Yahoo Finance

Na primeira iteração, notou-se que alguns ativos estão com último Caractere 'F' na lista, mas no Yahoo Finance havia ativos semelhantes sem este sufixo. Dessa forma, tentei sem o último caractere.

```python
stock_errors = []

for idx, data in df.iterrows():
    if not data['error']:
        continue
    
    ticker = data['Ativos'].strip()
    ticker = ticker[:-1]
    try:
        stock_data = scraper.get_data(data['Tipo'], ticker, from_date, to_date, interval)
        stock_data['Ticker'] = data['Ativos']
        stock_data['Tipo'] = data['Tipo']
        prices_data = pd.concat([prices_data, stock_data])
        print("log {}/{}: Success".format(idx,df.shape[0]))
    except (IOError, ExceptionsFacade.get_NoSuchElementException(), ExceptionsFacade.get_TimeoutException()) as er:
        stock_errors.append(idx)
        print("log {}/{}: Error -> {}".format(idx,df.shape[0], scraper.url))
df['error'] = 0
df['error'].iloc[stock_errors] = 1
```
